**AHP method**

```python
def normalizeMatrix(matrix):
    
    num_columns = len(matrix[0])
    column_sums = [sum(row[i] for row in matrix) for i in range(num_columns)]
    normalized_matrix = [[element/column_sums[col_index] for col_index, element in enumerate(row)] for row in matrix]

    return normalized_matrix
```

```python
norm_mat=normalizeMatrix(decision_matrix)
norm_mat
```

```python
def extract_weight_values(matrix):
    
    mat = normalizeMatrix(matrix)
    row_sums = [sum(row) for row in matrix]
    np_matrix = np.asarray(matrix)
    row_Averages = [element / np_matrix.shape[1] for element in row_sums]
    
    return row_Averages
```

```python
extract_weight_values(norm_mat)
```

```python
def final_score(matrix, altMatrices):
    
    weights_list = [extract_weight_values(element) for element in altMatrices]
    
    priority_matrix = np.array(weights_list).transpose()
    
    matrix_weights = np.array(extract_weight_values(matrix)).transpose()
    
    return np.dot(priority_matrix,matrix_weights)
    
```

```python
score = final_score(norm_mat,alternative_matrices)
score
```

```python
def rank(matrix):
    return sorted(list(matrix),reverse = True)

rank(score)
```

```python
def compute_CI(matrix):
    
    matrix_weights = np.array(extract_weight_values(matrix)).transpose()
    
    AW = [np.dot(row,matrix_weights) for row in matrix]
    
    Lambda = [aw / element for (aw,element) in zip(AW,matrix_weights)]
    
    m = max(Lambda)
    np_matrix = np.asarray(matrix)
    return (m-np_matrix.shape[0])/(np_matrix.shape[0]-1)
```

```python
compute_CI(norm_mat)
```

```python
def compute_CR(matrix, RI):
    
    CI = compute_CI(matrix)
    
    if RI !=0:
        CR = CI / RI 
    else:
        CR = 0
    return CR 
```

```python
def WASPASS1(wsm,wpm):
  lamb=0.5
  was=[]
  for s , p in zip(wsm, wpm):
    was.append(lamb*s+(1-lamb)*p)

  return was
```

```python
def discordance_matrix(dm, weights):
    num_alternatives, num_criteria = dm.shape
    D = np.zeros((num_alternatives, num_alternatives))
    for i in range(num_alternatives):
        for j in range(num_alternatives):
            if i == j:
                D[i,j] = 0
            else:
                discordance_values = []
                for k in range(num_criteria):
                    if dm[i,k] < dm[j,k]:
                        discordance_values.append(weights[k])
                if len(discordance_values) == 0:
                    D[i,j] = 0
                else:
                    D[i,j] = np.max(discordance_values) / np.sum(weights)
    return D
D = discordance_matrix(decision_matrix, weights)
D
```
